Analise da diferença de valores

```python
import pandas as pd

df_ontem= pd.read_csv("../data/raw/precos_computadores_2025-03-25.csv")

df_hoje= pd.read_csv("../data/raw/precos_computadores_2025-03-26.csv")

display(df_ontem)
display(df_hoje)
```

```python
#juntar as tabelas

df_comparado=pd.merge(df_ontem, df_hoje, on="nome", suffixes=("_ontem", "_hoje"))

display(df_comparado)
```

```python
#corrigir string para float

df_comparado["preco_ontem"] = (
    df_comparado["preco_ontem"]
    .str.replace("€", "")
    .str.replace(",", ".")
    .str.replace(" ", "")
    .str.strip()
    .astype(float)
)

df_comparado["preco_hoje"] = (
    df_comparado["preco_hoje"]
    .str.replace("€", "")
    .str.replace(",", ".")
    .str.replace(" ", "")
    .str.strip()
    .astype(float)
)

display(df_comparado)
```

```python
df_comparado["diferença"] = df_comparado["preco_hoje"] - df_comparado["preco_ontem"]

df_comparado["variação"] = df_comparado.apply(lambda
    row: "Subiu" if row["diferença"] > 0 
    else "Desceu" if row["diferença"] < 0 
    else "Igual", 
    axis=1
)

#ordena pelo maior aumento
df_comparacaogeral=df_comparado.sort_values(by="diferença", ascending=False).reset_index()
df_aumentaram=df_comparado[df_comparado["diferença"] > 0]  # aumentaram
df_diminuiram=df_comparado[df_comparado["diferença"] < 0]  # diminuíram


display(df_comparacaogeral)
display(df_aumentaram)
display(df_diminuiram)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

df_comparacaogeral["nome"] = df_comparacaogeral["nome"].str[:30]  # Limitar o nome dos produtos para melhor visualização

# Gerar gráfico de dispersão
plt.figure(figsize=(10, 6))

# Gráfico de dispersão para a diferença de preços
sns.scatterplot(x='nome', y='diferença', data=df_comparacaogeral, hue='variação', palette={"Subiu": "blue", "Desceu": "red", "Igual": "grey"}, s=100)

# Configurações de gráfico
plt.title('Variação de Preços de Computadores')
plt.xlabel('Nome do Produto')
plt.ylabel('Diferença de Preço (€)')
plt.xticks(rotation=90)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
